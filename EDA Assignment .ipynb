# Univariate analysis for categorical column:

```python
def plotfunc(df,col,title,hue=None):
    sns.set_style('dark')
    sns.set_context('poster')
    plt.rcParams['axes.labelsize']=13
    plt.rcParams['axes.titlesize']=15
    plt.rcParams['axes.titlepad']=28
    
    temp=pd.Series(data=hue)
    fig,ax =plt.subplots()
    width=len(df[col].unique()) +7 +4*len(temp.unique())
    fig.set_size_inches(width,8)
    plt.yscale('log')
    plt.title(title,fontsize=17)
    ax=sns.countplot(data=df,x=col,hue=hue,palette='bright')
    plt.setp(ax.get_xticklabels(),rotation=45, ha='right', rotation_mode='anchor')
    plt.show()
```

```python
#income range of Target 0 
plotfunc(Target0,col='AMT_INCOME_Range',title='Income Range of Target 0 ', hue='CODE_GENDER')
```

Male are more than female in range of 0-2L
